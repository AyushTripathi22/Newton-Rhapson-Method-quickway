##Algo for the code

```python
import math as m
def pra_nrm(f,df,ddf,x1,acc,maxitr):

  while abs(f(x1)*ddf(x1)/df(x1)/(df(x1))**2) > 1:
    print("the root does not lie in the given range,change the initial guess")
    x1=float(input("enter new x1:"))

  for i in range(maxitr):
    x0=x1-f(x1)/df(x1)

    if abs(x1-x0)< acc:
      break

    else:
      x1=x0

  print("the root of given function is %.4f"%x0)  
```

##Checking the code

```python
pra_nrm(lambda x:m.exp(x)-m.sin(x),lambda x:m.exp(x)-m.cos(x),lambda x: m.exp(x)+m.sin(x),-3,0.00001,1)
```
